```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("line graph")
plt.plot(x,y)
plt.show()
```

```python
x=[20,30,40,50]
mylabels=["Apples","bannas","cherries","dates"]
mycolors=["black","pink","orange","blue"]
myexplode=[0.3,0,0,0.5]
plt.pie(x,labels=mylabels,colors=mycolors,explode=myexplode)
plt.show()
```

```python
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
x=np.array(["g","f","d","s"])
y=np.array([3,4,7,10])
mycolor=np.array(["r","b","y","g"])
plt.barh(x,y,color=mycolor)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
mycolor=np.array(["r","b","y","g"])
sizes=[100,200,300,150]
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.scatter(x,y,c=mycolor,s=sizes,alpha=0.5,marker="*",edgecolor="black")
plt.show()
```
