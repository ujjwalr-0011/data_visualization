```python
import requests
from bs4 import BeautifulSoup
import re

def extract_images(soup):
    # Find all image tags
    images = soup.find_all('img',src=True)

    # Extract and print image URLs
    for img in images:
        print("Image Source:", img['src'])

def show_fonts(soup):
    # Find all linked CSS files
    css_links = soup.find_all('link', rel="stylesheet")

    # Extract font information from each CSS file
    for link in css_links:
        css_url = link['href']
        css_response = requests.get(css_url)

        if css_response.status_code == 200:
            # Parse CSS content to find font information
            font_urls = re.findall(r"url\((.*?)\)", css_response.text)
            print("Fonts used in", css_url)
            for font_url in font_urls:
                print("Font URL:", font_url)
        else:
            print(f"Failed to retrieve CSS file from {css_url}")

# URL of the website you want to scrape
url = "https://www.javatpoint.com/top-10-fastest-bikes-in-the-world"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract and print images
    print("\nImages:")
    extract_images(soup)

    # Show fonts used
    print("\nFonts:")
    show_fonts(soup)
else:
    print("Failed to retrieve the webpage")
```
